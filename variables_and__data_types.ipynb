```python
# Variables

first_name = 'Ess'
last_name = 'Kay'
full_name = 'Ess Kay'
country= 'Nigeria'
city = 'Remote'
age = 100
year = 2026
is_married = False
is_True = 'I am enjouing 30 Days of Python Programming'
is_light_on = True

# Multiple Variable

first_name, last_name, full_name, country, city, age, year = 'Ess', 'Kay', 'Ess Kay', 'Nigeria', 'Remote', 100, 2026
```

```python
# Exercises: Level 2

# Checking the data types of all my varibales

print(type(first_name))             # str
print(type(last_name))              # str
print(type(full_name))              # str
print(type(country))                # str
print(type(city))                   # str
print(type(age))                    # int
print(type(year))                   # int
print(type(is_married))             # bool
print(type(is_True))                # str
print(type(is_light_on))            # bool

# Using the len() built-in function to find the length the length of my first name

print(len(first_name))              # 8

# Comparing the lenght of my first name and last name

print(len(first_name)) == print(len(last_name))

# Declaring 5 as num_one and 4 as num_two

num_one = 5
num_two = 4

# Adding num_one and num_two and assign the value to a variable total

variable_total = num_one + num_two
print(variable_total)

# Subtracting num_two from num_one and assign the value to a varibale diff

variable_diff = num_one - num_two
print(variable_diff)

# Multiplying num_two and num_one and assign the value to a variable product

variable_product = num_one * num_two
print(variable_product)

# Divide num_one by num_two and assign the value to a variable division

variable_division = num_one / num_two
print(variable_division)

# Using modulus division to find num_two divided by num_one and assign the value to a variable remainder

variable_division = num_one % num_two
print(variable_division)

# Calculating num_one to the power of num_two and assign the value to a varibale exp

variable_exp = num_one ** num_two
print(variable_exp)

# Finding the floor divison of num_one by num_two and assign the value to a vairibale floor_division

variable_floor_division = num_one // num_two
print(variable_floor_division)

# Calculating the area of a circle and assign the value to a vairable name of area_of_circle

radius = 30
pi = 3.4
area_of_circle = pi * radius ** 2
print(area_of_circle)

# Calculating the circumference of a circle and assign the value to a variable name of circum_of_circle

circum_of_circle = 2 * pi * radius
print(circum_of_circle)

# Taking radius as user input and calculate the area

radius = float(input('What is the radius of the circle?'))
area_of_circle = pi * radius ** 2
print(area_of_circle)

# Using the input() built-in function to get first name, last name, country, and age ...

first_name = input('What is your first name?')
last_name = input('What is your last name?')
country = input('What is your country?')
age = input('What is your age?')

print(first_name)
print(last_name)
print(country)
print(age)
```

```python
# Python Control Flow

# Control Flow is a like the logic you write to control how your code runs, and ways to control the flow of your code includes a

# Conditional statement

# Loops - To repeat actions and tasks

# Tools to Control the Flow: Conditional Statements (if, else, & elif), and Loops (for & while)

# How to Control the flow of your code: break, pass, & continue

## Expressions to control the flow of your code

# Values: True & False

# Funtions: bool(), any(), all(), isinstance()

# Comparison Operators: ==, !=, <, >, >=, <=

# logical Operators: and, or, not

# Membership Operators: in, not in

# Identity Operator: is, is not
```

```python
# Boolean Expressions: This is used to control the flow of our codes and to build a logic in our conditional statements

# Example:

print(True)
print(False)
print((type)(True))
print(bool(123))
print(bool(("Hello")))
age = 0
print(bool(age))

email = "esskay@gmail.com"
phone_number = '09087654321'
username = ""

# Allows registration
# If any field is filled

print(any([email, phone_number, username])) # True becaue we have at least one information
print(all([email, phone_number, username])) # False because not all the list has information

# isinstance() - to check whether a value belongs to certain data type

print(isinstance(True, bool))
print(isinstance("ess", int))

# endswith and startwith: checks if the strings end or start with a specific word

print("Hello".endswith('o'))
print('hello'.startswith('4'))
```

```python
# Comparison Operators: to compare two values and return True or False based on the result

# Comparison Operators:

# == Equal to
# != Not Equal
# < Less than
# <= Less than or Equal
# > Greater than
# >= Greater than or Equal

# Examples

print(3 > 2) # Value

x_variable = 5

print(x_variable < 2) # Variable

print(2-1 != 2) # Expression

print(len('Esskay') == 3)

print(10 == 10)        # True because 10 is equal to 10
print(10 != 10)        # False because 10 is equal to 10 but the expression says it is not
print(7 > 3)           # True because 7 is greater than 3
print(10 >= 3)         # True because 10 is either greater than 3
print(3 <= 7)          # True because 3 is less than 7


## Note: You can compare strings too alphabetically, not just numbers

# Examples

print("a" < "b")        # True since Python treats strings alphabetically
print("a" == "b")       # False
print('a' == "A")       # False since Python is case sensitive

# Chain Comparison: To check multiple conditions in one line, just like in math (and it evaluates it from left to right, checking each condition one by one)

#example

print(1 < 4 < 6) # True

# practical examples: Is age between 18 and 30?

age = 20

print(18 <= age <=30)
```

```python
# Logical Operators (and | or): Used to combine multiples boolean expressions

print(3 > 1 and 5 < 1)
print(3 < 1 or 5 < 1)

print(3 > 1 or 5 < 1)
print(3 < 1 or 5 < 1)

#checks if the system is under pressure

cpu_usage = 70
memory_usage = 50

print(cpu_usage > 90 or memory_usage > 90)

#checking user credentials before login
email = True
password = True

print(email and password)

#Logican Operators (NOT) - flips the original results

print(not 3 > 2) # False because 3 is greater than 2
print(not 3 < 2) # True
print(not True)
print(not False)

name = ""
print(not name)
print(not 0)

# Execution Order: "and" has higher priority than "or" which means and will be excuted first
# Parenthese () first to control order

# Control Mixed Conditions

# Example
# Allow access only if the user is logged in
# or they are a guest
# but they must not be banned

is_logged_in = True
is_guest = False
is_banned = False

print((is_logged_in or is_guest) and not is_banned)
```
